```python
#K-Means Clustering:

# Select features
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]
```

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)
```

```python
# Display the clustered DataFrame
print(df)
```

```python
#Principal Component Analysis (PCA):

# Select features
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
components = pca.fit_transform(X)
```

```python
# Create a new DataFrame with the principal components
df_pca = pd.DataFrame(data=components, columns=['PC1', 'PC2'])
print(df_pca)
```

```python
#Independent Component Analysis (ICA):

# Select features
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]
```

```python
# Apply ICA for independent component analysis
ica = FastICA(n_components=3)
components = ica.fit_transform(X)
```

```python
# Create a new DataFrame with the independent components
df_ica = pd.DataFrame(data=components, columns=['IC1', 'IC2', 'IC3'])
print(df_ica)
```

```python
#DBSCAN:

#library for it
from sklearn.cluster import DBSCAN

features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['cluster'] = dbscan.fit_predict(X)
print(df['cluster'])
```

```python
#t-Distributed Stochastic Neighbor Embedding (t-SNE):

#library for it
from sklearn.manifold import TSNE

# Assuming df is your DataFrame with features
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]


# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2)
components = tsne.fit_transform(X)
df_tsne = pd.DataFrame(data=components, columns=['TSNE1', 'TSNE2'])
print(df_tsne)
```
